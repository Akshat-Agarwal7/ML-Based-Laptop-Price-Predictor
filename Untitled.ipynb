```python
def oss(name):
    name=name.lower()
    if 'windows' in name:
        return 'Windows'
    elif 'mac' in name:
        return 'Mac'
    else:
        return "Other OS"
```

```python
df['OpSys']=df['OpSys'].apply(oss)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
categorical_cols = ['Company','TypeName','chipset','Gpu_brand','OpSys']

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough'
)
```

```python
pipe = Pipeline([
    ('preprocess', preprocessor),
    ('model', LinearRegression())
])
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_p=pipe.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test, y_p)
r2
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error

# Categorical columns in your dataset
categorical_cols = ['Company', 'TypeName', 'chipset', 'Gpu_brand', 'OpSys']

# Step 1: Preprocessor
step1 = ColumnTransformer(
    transformers=[
        ('col_tnf', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough'
)

# Step 2: Random Forest model (same hyperparameters as image)
step2 = RandomForestRegressor(
    n_estimators=100,
    random_state=3,
    max_samples=0.5,
    max_features=0.75,
    max_depth=15,
    n_jobs=-1
)

# Pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Train
pipe.fit(X_train, Y_train)

# Predict
y_pred = pipe.predict(X_test)

# Evaluation
print("R2 score:", r2_score(Y_test, y_pred))
print("MAE:", mean_absolute_error(Y_test, y_pred))
```
