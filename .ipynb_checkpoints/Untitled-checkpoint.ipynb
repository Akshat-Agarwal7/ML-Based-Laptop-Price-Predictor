# EDA

-we will first drop unnamed             
-in RAM, we will remove GB        
-from weight we will remove kg

remoing netbook because they are low in samples and they are obsolete and no longer common in the current market.

```python
df = df[df['TypeName'] != 'Netbook']
```

```python
df.head()
```

CPU

```python
df['Cpu'].value_counts()
```

```python
df['CPU name']=df['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
df.head()
```

```python
def chipset(text):
    if text=='Intel Core i5' or text=='Intel Core i7' or text=='Intel Core i3':
        return text
    else:
        if text.split()[0]=='Intel':
            return("different intel processor")
        else:
            return('AMD processor')
```

```python
df['chipset']=df['CPU name'].apply(chipset)
```

```python
df.head()
```

```python
df['chipset'].value_counts().plot(kind='bar')
```

```python
sns.barplot(x=df['chipset'],y=df['Price'])
plt.xticks(rotation="vertical")
plt.show()
```

```python
df['ips'] = df['ScreenResolution'].str.contains('IPS', case=False).map({True: 'Yes', False: 'No'})
```

```python
df.head()
```

```python
sns.barplot(x=df['ips'],y=df['Price'])
```

we have resolution and inches, from them wwe can compute ppi and remove those columns

```python
df[['X_res', 'Y_res']] = df['ScreenResolution'].str.extract(r'(\d+)x(\d+)').astype(int)
```

```python
df['ppi']=np.sqrt(df['X_res']**2+df['Y_res']**2)/df['Inches']
```

```python
sns.scatterplot(x=df['ppi'],y=df['Price'])
```
